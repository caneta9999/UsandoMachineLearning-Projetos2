```python
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go
```

```python
dados = pd.read_csv('data_pertanian_modern_fake.csv').drop(['windspeed','precipitation','solar irradiance'],axis=1)
dados
```

# KMeans

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=3)
kmeans.fit(dados)
```

```python
grafico1 = px.scatter(x = dados['humidity'],y= dados['temperature'], color = kmeans.labels_)
grafico2 = px.scatter(x = kmeans.cluster_centers_[:,0], y = kmeans.cluster_centers_[:,1], size = [25, 25, 25])
go.Figure(data = grafico1.data + grafico2.data).show()
```

```python
from sklearn.metrics import silhouette_score
```

```python
min = 2
max = 11
```

```python
kmeansMeasures = [-1,0] #id best silhouette, silhouette score
silhouetteScores = []
for i in range(min,max):
  kmeans = KMeans(n_clusters=i)
  kmeans.fit(dados)
  silhouetteScore = silhouette_score(dados, kmeans.labels_)
  silhouetteScores.append(silhouetteScore)
  if kmeansMeasures[0] == -1:
    kmeansMeasures[0] = i
    kmeansMeasures[1] = silhouetteScore
  elif kmeansMeasures[1] < silhouetteScore:
    kmeansMeasures[0] = i
    kmeansMeasures[1] = silhouetteScore
```

```python
kmeansMeasures
```

```python
silhouetteScores
```

```python
grafico1 = px.line(x =range(min,max) , y = silhouetteScores)
go.Figure(data = grafico1.data)
```

# MeanShift
